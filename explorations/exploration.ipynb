# EDA

```python
import requests
import json
import pandas as pd

with open("data/brottsbalken.json", encoding="utf-8") as f:
    data = json.load(f)

# df = pd.json_normalize(data)
# df = pd.DataFrame(data)

df = pd.json_normalize(
    data["kapitel"],
    record_path="paragraphs",
    meta=["chapter_number", "title"],
    errors="ignore"
)
```

## Cleaning the data

```python
df.drop(columns=["repealed"], inplace=True)
```

## Remove chapter number and 'kap' from title

```python
df["title"] = df["title"].str.replace(r'^(\d+\s*)?(kap\.\s*)?(\d+\s*)?', '', regex=True)
```

```python
df
```

```python
df["chapter_number"] = df["chapter_number"].astype(str) + " kap."
```

```python
df.to_json(
    "../data/brottsbalken_clean.json",
    orient="records",
    indent=4,
    force_ascii=False
)
```
