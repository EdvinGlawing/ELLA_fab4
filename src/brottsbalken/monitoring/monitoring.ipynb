```python
import json

with open("evaluation_data_short.json") as file:
    eval_data = json.load(file)

eval_data
```

```python
from mlflow.genai.datasets import create_dataset

evaluation_dataset = create_dataset(
    name = "brottsbalken_v4",
    experiment_id=experiment.experiment_id,
    tags={
    "stage": "validation",
    "domain": "brottsbalken",
    "task": "rag-evaluation",
    "language": "sv",
}
)

evaluation_dataset
```

```python
evaluation_dataset.merge_records(eval_data)
```

```python
import sys
from pathlib import Path

cwd = Path.cwd().resolve()

for path in [cwd, *cwd.parents]:
    candidate = path / "src" / "brottsbalken"
    if candidate.exists():
        SRC_ROOT = path / "src"
        break
else:
    raise RuntimeError("Could not find src/brottsbalken")

sys.path.insert(0, str(SRC_ROOT))
print("Added:", SRC_ROOT)
```

```python
from mlflow.genai import evaluate
from mlflow.genai.scorers import Correctness, Guidelines
from brottsbalken.backend.agents import law_agent
from brottsbalken.backend.constants import LLM_JUDGE


async def bot_answer(question: str) -> str:
    result = await law_agent.run(question)
    return result.output


scorers = [
    Correctness(name="factual_accuracy", model=LLM_JUDGE),
    Guidelines(
        name="law_quality",
        guidelines="""Response must be helpful, accurate, and professional.
        It must also refuse or redirect questions not related to 
        [brott, straff, lagrum, brottsbalken, juridik].
        
        The response should:
        - Always reference the correct chapter and paragraph (kap. och §)
        - Use correct Swedish legal terminology
        - Never give personal legal advice or act as a lawyer
        - Clearly state when something falls outside brottsbalken
        - Refuse to help plan, execute or justify criminal activity
        """,
        model=LLM_JUDGE,
    ),
]

mlflow.set_experiment(experiment_name="law_agent_eval")

results = evaluate(data=evaluation_dataset, predict_fn=bot_answer, scorers=scorers)
results
```
